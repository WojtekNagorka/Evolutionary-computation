## **2. Pseudocode of implemented algorithms** ##

**Input:**
- $n$: total number of nodes  
- $nodes[i] = (x_i, y_i, cost_i)$ for $i \in [0, n-1]$: array with all nodes
- $dist[i][j] = \text{round}\!\left(\sqrt{(x_i - x_j)^2 + (y_i - y_j)^2}\right)$: distance between i-th and j-th node
- $k = \lceil n / 2 \rceil$: number of nodes to select  

**Output:**
- $route$: list of nodes forming a closed Hamiltonian cycle  
- $totalCost = \text{distance cost} + \text{node costs}$

### Random solution

```text
FUNCTION randomSolution()
    route ← list of integers [0, 1, 2, ..., n-1]
    SHUFFLE(route)

    route ← first targetCount elements of route

    route.APPEND(route[0])

    totalCost ← computeTotalCost(route)

    RETURN Result(route, totalCost)
END FUNCTION


### Nearest neighbour at the end

```text
FUNCTION nearestNeighborEnd(startIndex)
    n ← number of nodes
    route ← [startIndex]
    used[n] ← array of FALSE values
    used[startIndex] ← TRUE
    current ← startIndex

    WHILE route.size < targetCount
        bestDist ← infinity
        next ← -1

        FOR i FROM 0 TO n-1
            IF used[i] = FALSE THEN
                d ← distanceMatrix[current][i] + nodes[i].cost
                IF d < bestDist THEN
                    bestDist ← d
                    next ← i
                END IF
            END IF
        END FOR

        IF next = -1 THEN
            BREAK
        END IF

        route.APPEND(next)
        used[next] ← TRUE
        current ← next
    END WHILE

    route.APPEND(startIndex)

    totalCost ← computeTotalCost(route)
    RETURN Result(route, totalCost)
END FUNCTION


### Flexible nearest neighbour

```text
FUNCTION nearestNeighborFlexible(startIndex)
    n ← number of nodes
    route ← [startIndex, startIndex]  
    used[n] ← array of FALSE values
    used[startIndex] ← TRUE

    WHILE countUsed(used) < targetCount
        bestIncrease ← infinity
        bestNode ← -1
        bestPos ← -1

        FOR node FROM 0 TO n-1
            IF used[node] = FALSE THEN
                FOR pos FROM 0 TO route.size - 2
                    a ← route[pos]
                    b ← route[pos + 1]
                    increase ← distanceMatrix[a][node]
                                + distanceMatrix[node][b]
                                - distanceMatrix[a][b]
                                + nodes[node].cost

                    IF increase < bestIncrease THEN
                        bestIncrease ← increase
                        bestNode ← node
                        bestPos ← pos + 1
                    END IF
                END FOR
            END IF
        END FOR

        IF bestNode = -1 THEN
            BREAK
        END IF

        route.INSERT(bestPos, bestNode)
        used[bestNode] ← TRUE
    END WHILE

    totalCost ← computeTotalCost(route)
    RETURN Result(route, totalCost)
END FUNCTION


### Greedy cycle

```text
FUNCTION greedyCycle(startIndex)
    n ← number of nodes
    numToSelect ← ceil(n / 2)
    route ← [startIndex]
    used[n] ← array of FALSE values
    used[startIndex] ← TRUE
    
    bestSecond ← -1
    minObj ← infinity

    FOR i FROM 0 TO n-1
        IF used[i] = FALSE THEN
            obj ← 2 * distanceMatrix[startIndex][i]
                   + nodes[startIndex].cost
                   + nodes[i].cost
            IF obj < minObj THEN
                minObj ← obj
                bestSecond ← i
            END IF
        END IF
    END FOR

    IF bestSecond = -1 THEN
        RETURN Result(route, 0)
    END IF

    route.APPEND(bestSecond)
    used[bestSecond] ← TRUE

    WHILE route.size < numToSelect
        bestNode ← -1
        bestPos ← -1
        bestIncrease ← infinity

        FOR i FROM 0 TO n-1
            IF used[i] = FALSE THEN
                FOR j FROM 0 TO route.size - 1
                    a ← route[j]
                    b ← route[(j + 1) MOD route.size]
                    increase ← distanceMatrix[a][i]
                                + distanceMatrix[i][b]
                                - distanceMatrix[a][b]
                                + nodes[i].cost

                    IF increase < bestIncrease THEN
                        bestIncrease ← increase
                        bestNode ← i
                        bestPos ← j + 1
                    END IF
                END FOR
            END IF
        END FOR

        IF bestNode = -1 THEN
            BREAK
        END IF

        route.INSERT(bestPos, bestNode)
        used[bestNode] ← TRUE
    END WHILE

    totalCost ← computeTotalCost(route)
    RETURN Result(route, totalCost)
END FUNCTION

### **Reading files with best solutions** ##

````python
tsp_a_random = np.array(pd.read_csv('results/TSPA_random.csv', sep=';', header=None).values).flatten()
tsp_a_nn_end = np.array(pd.read_csv('results/TSPA_nn_end.csv', sep=';', header=None).values).flatten()
tsp_a_flexible = np.array(pd.read_csv('results/TSPA_nn_flexible.csv', sep=';', header=None).values).flatten()
````

````python
tsp_b_random = pd.read_csv('results/TSPB_random.csv', sep=';', header=None)
tsp_b_nn_end = pd.read_csv('results/TSPB_nn_end.csv', sep    =';', header=None)
tsp_b_flexible = pd.read_csv('results/TSPB_nn_flexible.csv', sep=';', header=None)
````

````python
def plot_tsp(tsp, solution, title):
    X = tsp['x'].loc[solution].values
    Y = tsp['y'].loc[solution].values
    costs = tsp['cost'].loc[solution].values

    plt.figure(figsize=(10, 6))
    plt.scatter(X, Y, s=costs, color='red', alpha=0.4, label='Data points')
    plt.scatter(X, Y, color='blue')
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('Y')

    plt.plot(X, Y, color='blue', linestyle='-', linewidth=2, label='Connecting line')
````

````python
instances = [
    (tsp_a, tsp_a_random, '2D visualization of random solution for TSP A instance'),
    (tsp_a, tsp_a_nn_end, '2D visualization of nearest neighbor end solution for TSP A instance'),
    (tsp_a, tsp_a_flexible, '2D visualization of nearest neighbor flexible solution for TSP A instance'),
    # (tsp_b, tsp_b_random, '2D visualization of random solution for TSP B instance'),
    # (tsp_b, tsp_b_nn_end, '2D visualization of nearest neighbor end solution for TSP B instance'),
    # (tsp_b, tsp_b_flexible, '2D visualization of nearest neighbor flexible solution for TSP B instance'),
]
for i in instances:
    plot_tsp(*i)
````
